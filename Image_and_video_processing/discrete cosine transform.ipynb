Aim: To us discrete cosine transform to compress the given image.

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt 
from skimage.color import rgb2gray
import cv2
import scipy
from scipy.fftpack import dct
from scipy.fftpack import idct
```

```python
img=io.imread('/content/cat (1).png')
img=rgb2gray(img)
img=255*img
```

```python
img[2][3]
```

```python
[rows,cols]=img.shape
```

```python
N=8
image_dct=np.zeros((rows,cols),dtype=int)
for r in range(rows//N):
  for c in range(cols//N):
    temp=dct(img[r*N:(r+1)*N,c*N:(c+1)*N])
    temp[2:N,2:N]=0
    image_dct[r*N:(r+1)*N,c*N:(c+1)*N]=temp
```

```python
image_idct=np.zeros((rows,cols),dtype=int)
for r in range(rows//N):
  for c in range(cols//N):
    temp=idct(image_dct[r*N:(r+1)*N,c*N:(c+1)*N])
    image_idct[r*N:(r+1)*N,c*N:(c+1)*N]=temp
```

```python
plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(img, cmap = 'gray')
plt.title ('Original Image')
plt.subplot(1,3,2)
plt.imshow(image_dct, cmap = 'gray')
plt.title ('Image after DCT')
plt.subplot(1,3,3)
plt.imshow(image_idct, cmap = 'gray')
plt.title ('Inverse of Image after DCT')
```

Conclusion: DCT transform of the given image is calculated. Inverse of the DCT transform of the above image shows the original image. If last 6 rows and 6 cols of 8x8 DCT coefficients are converted to 0 then the inverse transform of the image shows pixelated image. If less no of rows and cols are converted to 0 then pixelation effect is less than the earlier one.
