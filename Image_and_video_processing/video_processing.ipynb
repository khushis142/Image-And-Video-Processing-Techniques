Aim: To process images of the given video and convert processed images to a new video

```python
import matplotlib.pyplot as plt 
import cv2 #cv = computer vision
```

```python
cap=cv2.VideoCapture('/content/nature.mp4')
i=0
while cap.isOpened():
  ret,fr=cap.read()
  if ret==False:
    break
  cv2.imwrite('video_f'+str(i)+'.jpg',fr)
  i=i+1
```

```python
num_frames=i
num_frames
```

```python
image=cv2.imread('/content/video_f76.jpg',0) #put 0 when it is a gray image and 1 when its a coloured image
[rows,cols]=image.shape
[rows,cols]
```

```python
strt_fr=int(num_frames*0.5)
stp_fr=int(num_frames*0.8)
(strt_fr,stp_fr)
```

```python
strt_rw=int(rows*0.10)
stp_rw=int(rows*0.60)
strt_col=int(cols*0.10)
stp_col=int(cols*0.60)
(strt_rw,stp_rw,strt_col,stp_col)
```

```python
for i in range(strt_fr,stp_fr):
  i_frame=cv2.imread('video_f'+str(i)+'.jpg',0)
  i_frame[strt_rw:stp_rw,strt_col:stp_col]=255
  cv2.imwrite('video_f'+str(i)+'.jpg',i_frame)
```

```python
cap.release()
cv2.destroyAllWindows()
```

```python
import glob
```

```python
i_array=[]
for f in glob.glob('video_f*.jpg'):
  image=cv2.imread(f)
  i_array.append(image)
```

```python
size=[cols,rows]
size
```

```python
video_out=cv2.VideoWriter('new.avi',cv2.VideoWriter_fourcc(*'DIVX'),10,size) # new video with name new.avi created coded is 4cc DIVX 10 frames per sec of size=[width, height] so earlier we took size as [cols, rows]
ln=len(i_array)
for i in range(ln):
  video_out.write(i_array[i])
video_out.release()
```

Conclusion: All the frames of the given video are extracted. Some of the frames are processed to add white patches. New video is generate by using oringinal and processed frames. New video shows white patches in some of the frames
