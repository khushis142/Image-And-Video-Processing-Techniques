```python
from skimage import io #to read image into image matrix. io has different options like read, write etc
import matplotlib.pyplot as plt # to show image matrix as image on screen
from skimage.color import rgb2gray #to convert color image to gray image
```

```python
im=io.imread('/content/lenna_grey (1).jpg') #to read image
im.shape #Give image matrix dimensions[rows,columns,no of matrix(3 for rgb)]
```

```python
im_gray=rgb2gray(im) #to convert color image to gray image
im_gray.shape #3 matrix rgb converted to 1 matrix (gray image) after some processing


```

```python
plt.figure() #creates new space for each subplot so that subsequent subplots 
plt.subplot(1,2,1) #https://www.w3schools.com/python/matplotlib_subplot.asp
plt.imshow(im, cmap='gray') #c map is color map. cmap=gray shows min pixel intensity as black and max pixel intensity as white. if cmap not given the default is shown in blue to yellow intensity range
#https://www.geeksforgeeks.org/matplotlib-pyplot-imshow-in-python/
plt.subplot(1,2,2)
plt.imshow(im_gray,cmap='gray')
```

```python
im_gray[2][3] #shows normalised pixel intensity of pixel in 2nd row and 3rd column
```

```python
im_gray=im_gray*255 #multiplies normalised intensity of all pixels of image by 255 so we get actual intensity of 8 bit image
im_gray[2][3]
```

```python
[rw,cl]=im_gray.shape
print([rw,cl])
```

Power law transformation: intensity of every pixel of image raised to gamma 

```python
image_less=im_gray.copy() #copy of im_gray made in image_less(variable name since we want gamma < 1). original image is retained
```

```python
gamma=0.8
for r in range(rw):
  for c in range(cl):
    temp1=im_gray[r][c]
    temp2=temp1**gamma
    image_less[r][c]=temp2
```

```python
plt.figure()
plt.subplot(1,3,1)
plt.imshow(im_gray,cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(image_less,cmap='gray')
plt.title('gamma < 1')
```

```python
image_more=im_gray.copy()
gamma=1.2
for r in range(rw):
  for c in range(cl):
    temp1=im_gray[r][c]
    temp2=temp1**gamma
    image_more[r][c]=temp2
```

```python
plt.figure()
plt.subplot(1,3,1)
plt.imshow(im_gray,cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(image_less,cmap='gray')
plt.title('gamma < 1')
plt.subplot(1,3,3)
plt.imshow(image_more,cmap='gray')
plt.title('gamma > 1')
```

Thresholding


```python
im2=io.imread('/content/watch (1).png')
im2_gray=rgb2gray(im2)
im2_gray.shape
```

```python
plt.figure() 
plt.subplot(1,2,1) 
plt.imshow(im2, cmap='gray') 
plt.subplot(1,2,2)
plt.imshow(im2_gray,cmap='gray')
```

```python
[rw2,cl2]=im2_gray.shape
print([rw2,cl2])
```

```python
im2_gray=im2_gray*255
im2_gray[2][3]
```

```python
th=100
im_th1=im2_gray.copy()
for r in range(rw2):
  for c in range(cl2):
    if im_th1[r][c]>th:
      im_th1[r][c]=255
    else:
      im_th1[r][c]=0
```

```python
im_th2=im2_gray.copy()
for r in range(rw2):
  for c in range(cl2):
    if im_th2[r][c]>th:
      im_th2[r][c]=255
    
```

```python
plt.figure()
plt.subplot(1,3,1)
plt.imshow(im2_gray,cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(im_th1,cmap='gray')
plt.title('Thresholding without background')
plt.subplot(1,3,3)
plt.imshow(im_th2,cmap='gray')
plt.title('Thresholding with background')
```

Power transformation is applied on the given test image for gamma = 0.8.
The image turns brighter and for gamma = 0.5 brighter than for gamma = 0.8.
For gamma = 1.5 the image turns darker

Negative operation on the test image changes range of black to white pixels to white to black pixels

Thresholding is applied to the test image for threshold of 100. All pixels with intensity > 100 are highlighted ( made white ) and remaining pixels are left unchanged or made to zero. Thesholding can used to change gray scale image to binary image.
