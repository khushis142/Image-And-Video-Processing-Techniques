```python
from skimage import io #to read image into image matrix. io has different options like read, write etc
import matplotlib.pyplot as plt # to show image matrix as image on screen
from skimage.color import rgb2gray #to convert color image to gray image
import numpy as np #NumPy is a Python library used for working with arrays. It also has functions for working in domain of linear algebra, fourier transform, and matrices.
```

```python
im=io.imread('/content/Unequalized_image.jpg') #to read image
im.shape #Give image matrix dimensions[rows,columns,no of matrix(3 for rgb)]
```

```python
im_gray=rgb2gray(im) #to convert color image to gray image
im_gray.shape #3 matrix rgb converted to 1 matrix (gray image) after some processing
```

```python
im_gray[2][3] #shows normalised pixel intensity of pixel in 2nd row and 3rd column
```

```python
im_gray=im_gray*255 #multiplies normalised intensity of all pixels of image by 255 so we get actual intensity of 8 bit image
im_gray[2][3]
```

```python
[rows,cols]=im_gray.shape
(rows,cols)
```

```python
#histo_image is a variable for no of pixels for each intensity in image. bins is diff intensities i.e 0-255
#Compute the histogram of a dataset.
#im.flatten to convert the histogram (values of intensity of all pixels over flattened(1 dimension) array)
#256 gives no of bins of equal range
histo_image,bins=np.histogram(im_gray.flatten(),256)
```

```python
#print(histo_image)
```

```python
#print(bins)
```

```python
cum = histo_image.cumsum() #for cumulative frequency i.e nr
cum_nex = cum*255/(rows*cols) #new frequency it should be i.e s
new_intensity = cum_nex.astype(int) #s as integer
```

```python
im_eq = im_gray.copy() #im_eq is image to be equalised
```

```python
#print(cum)
```

```python
#print(new_intensity)
```

```python
for i in range (rows):
  for j in range (cols):
    original_pixel = im_gray[i][j]
    original_pixel = original_pixel.astype(int)
    s=new_intensity[original_pixel] #
    im_eq[i][j] = s
```

```python
plt.figure(figsize=(7,7))

plt.subplot(2,2,1)

plt.imshow(im_gray, cmap = 'gray')

plt.title ('Original Image')

plt.subplot(2,2,2)

plt.hist(im_gray.ravel(),bins=256)

plt.title ('Original Image Histogram')
```

```python
plt.figure(figsize=(7,7))

plt.subplot(2,2,3)

plt.imshow(im_eq, cmap = 'gray')

plt.title ('Equalised Image')

plt.subplot(2,2,4)

plt.hist(im_eq.ravel(),bins=256)

plt.title ('Equalised Image Histogram')
```
